P2. RFM preliminar


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el archivo CSV
df = pd.read_csv("clientes_segmentos (1).csv")

# Verificar nombres de las columnas
print(df.columns.tolist())  # Esto nos ayudará a ver qué nombre tiene la columna de monto

# Asegúrate de que la columna 'fecha_ultima_compra' esté en formato datetime
df['fecha_ultima_compra'] = pd.to_datetime(df['fecha_ultima_compra'])

# Fecha de referencia (última fecha de compra + 1 día)
fecha_ref = df['fecha_ultima_compra'].max() + pd.Timedelta(days=1)

# Agrupar por cliente y calcular Recency, Frequency y Monetary
rfm = df.groupby('id_cliente').agg({
    'fecha_ultima_compra': lambda x: (fecha_ref - x.max()).days,  # Recency
    'id_cliente': 'count',                                        # Frequency
    'total_gastado': 'sum'                                        # Monetary (Total gastado)
})

# Renombrar columnas a RFM
rfm.columns = ['Recency', 'Frequency', 'Monetary']
rfm = rfm.reset_index()

# Mostrar los primeros registros de RFM
print(rfm.head())

# Matriz de correlación
corr = rfm[['Recency', 'Frequency', 'Monetary']].corr()

# Visualización del heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlación entre Recency, Frequency y Monetary')
plt.show()
```

La combinación de Recency y Frequency es la que captura mejor el valor futuro del cliente, ya que muestra tanto el compromiso reciente como el nivel de lealtad a largo plazo.

**Métricas de Negocio (CAC, LTV, ROI)**

P3. CAC detallado

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar el archivo CSV
df = pd.read_csv("clientes_segmentos (1).csv")

# Asegurarse de que 'costo_adquisicion' esté en formato numérico
df['costo_adquisicion'] = pd.to_numeric(df['costo_adquisicion'], errors='coerce')

# Calcular número de clientes activos por plataforma
clientes_por_plataforma = df.groupby('plataforma')['id_cliente'].nunique()

# Calcular el porcentaje de clientes activos por plataforma
clientes_totales = df['id_cliente'].nunique()
porcentaje_activos = (clientes_por_plataforma / clientes_totales) * 100

# Calcular el CAC por plataforma
cac_por_plataforma = df.groupby('plataforma')['costo_adquisicion'].sum() / clientes_por_plataforma

# Calcular el CAC ponderado
cac_ponderado = cac_por_plataforma * (porcentaje_activos / 100)

# Crear un DataFrame con los resultados
resultados = pd.DataFrame({
    'CAC Ponderado': cac_ponderado,
    'Clientes Activos (%)': porcentaje_activos
})

# Graficar el resultado en barras apiladas
resultados.plot(kind='bar', stacked=True, figsize=(10, 6))

# Añadir etiquetas y título
plt.xlabel('Plataforma')
plt.ylabel('Valor')
plt.title('CAC Ponderado y Porcentaje de Clientes Activos por Plataforma')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

P4. LTV proyectado


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el archivo CSV
df = pd.read_csv("clientes_segmentos (1).csv")

# Asegurarse de que las columnas necesarias estén en formato numérico
df['costo_adquisicion'] = pd.to_numeric(df['costo_adquisicion'], errors='coerce')
df['total_gastado'] = pd.to_numeric(df['total_gastado'], errors='coerce')
df['compras'] = pd.to_numeric(df['compras'], errors='coerce')

# Calcular promedio ticket (total gastado dividido por el número de compras)
df['promedio_ticket'] = df['total_gastado'] / df['compras']

# Calcular la frecuencia mensual (número de compras / meses de actividad)
# Supongamos que cada cliente ha estado activo desde la fecha más antigua hasta la fecha más reciente
df['fecha_ultima_compra'] = pd.to_datetime(df['fecha_ultima_compra'])
fecha_maxima = df['fecha_ultima_compra'].max()
df['dias_actividad'] = (fecha_maxima - df['fecha_ultima_compra']).dt.days
df['frecuencia_mensual'] = df['compras'] / (df['dias_actividad'] / 30)  # Convertimos los días a meses

# Asignar tasa de retención según el intervalo de días desde la última compra
def obtener_retencion(dias):
    if dias <= 90:
        return 0.80  # 80% retención
    elif dias <= 180:
        return 0.60  # 60% retención
    else:
        return 0.30  # 30% retención

df['retencion'] = df['dias_actividad'].apply(obtener_retencion)

# Calcular el LTV
df['LTV'] = df['promedio_ticket'] * df['frecuencia_mensual'] * 12 * df['retencion']

# Graficar la distribución del LTV (histograma + KDE)
plt.figure(figsize=(10, 6))
sns.histplot(df['LTV'], kde=True, bins=30)
plt.title('Distribución de LTV')
plt.xlabel('LTV')
plt.ylabel('Frecuencia')
plt.show()

# Reportar los clientes en el percentil 90 de LTV
percentil_90 = np.percentile(df['LTV'], 90)
clientes_percentil_90 = df[df['LTV'] >= percentil_90]

print(f"El valor del LTV en el percentil 90 es: {percentil_90}")
print(f"Clientes en el percentil 90 del LTV:")
print(clientes_percentil_90[['id_cliente', 'LTV']])
```

P5. Margen neto

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar el archivo CSV
df = pd.read_csv("clientes_segmentos (1).csv")

# Asegurarse de que las columnas necesarias sean numéricas
df['total_gastado'] = pd.to_numeric(df['total_gastado'], errors='coerce')
df['compras'] = pd.to_numeric(df['compras'], errors='coerce')
df['costo_adquisicion'] = pd.to_numeric(df['costo_adquisicion'], errors='coerce')
df['fecha_ultima_compra'] = pd.to_datetime(df['fecha_ultima_compra'])

# Calcular promedio_ticket
df['promedio_ticket'] = df['total_gastado'] / df['compras']

# Calcular frecuencia_mensual: Compras / meses de actividad
df['dias_actividad'] = (df['fecha_ultima_compra'].max() - df['fecha_ultima_compra']).dt.days
df['frecuencia_mensual'] = df['compras'] / (df['dias_actividad'] / 30)  # Convertimos días a meses

# Calcular retención basada en días desde la última compra
df['retencion'] = df['dias_actividad'].apply(lambda x: 0.80 if x <= 90 else (0.60 if x <= 180 else 0.30))

# Calcular LTV
df['LTV'] = df['promedio_ticket'] * df['frecuencia_mensual'] * 12 * df['retencion']

# Calcular CAC por plataforma
clientes_por_plataforma = df.groupby('plataforma')['id_cliente'].nunique()
cac_por_plataforma = df.groupby('plataforma')['costo_adquisicion'].sum() / clientes_por_plataforma

# Asignar CAC por plataforma a cada cliente
df['CAC'] = df['plataforma'].map(cac_por_plataforma)

# Crear columna margen (LTV - CAC)
df['margen'] = df['LTV'] - df['CAC']

# Graficar el boxplot comparando margen por plataforma
plt.figure(figsize=(10, 6))
sns.boxplot(x='plataforma', y='margen', data=df)
plt.title('Margen (LTV - CAC) por Plataforma')
plt.show()

# Plataforma con el mayor margen promedio
margen_promedio = df.groupby('plataforma')['margen'].mean()
mejor_plataforma = margen_promedio.idxmax()
print(f"La plataforma con mejor rentabilidad incremental es: {mejor_plataforma}")
print(f"Margen promedio por plataforma:\n{margen_promedio}")
```

**Segmentación y Modelado**

P6. Selección de k (K–Means) = 3


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Cargar el archivo CSV
df = pd.read_csv("clientes_segmentos (1).csv")

# Selección de variables para clustering
variables = ['edad', 'ingresos', 'total_gastado', 'compras']  # O las columnas que necesites

# Estándarizar las variables
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df[variables])

# Aplicar K-Means con k=3
kmeans = KMeans(n_clusters=3, random_state=42)
df['cluster'] = kmeans.fit_predict(df_scaled)

# Mostrar los centros de los clusters
centroids = pd.DataFrame(kmeans.cluster_centers_, columns=variables)
print("Centros de los clusters:\n", centroids)

# Visualizar los clusters en un gráfico de dispersión 2D
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['edad'], y=df['ingresos'], hue=df['cluster'], palette='viridis', s=100, alpha=0.6)
plt.title('Clusters de Clientes (K=3)')
plt.xlabel('Edad')
plt.ylabel('Ingresos')
plt.show()

# Graficar el método del codo (opcional) para validar la elección de k=3
# Usamos este gráfico solo para referencia
inertia = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df_scaled)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), inertia, marker='o')
plt.title('Método del Codo para Selección de k')
plt.xlabel('Número de Clusters')
plt.ylabel('Inercia')
plt.show()
```

P7. Segmentación final

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Cargar el archivo CSV
df = pd.read_csv("clientes_segmentos (1).csv")

# Verificar las columnas para identificar las correctas
print(df.columns)

# Seleccionar las columnas correctas para el clustering
# Asegúrate de que estas columnas existan en el DataFrame
variables = ['edad', 'ingresos', 'interacciones', 'compras']  # Asegúrate de que estas columnas existan en tu archivo CSV

# Normalizar las variables seleccionadas
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df[variables])

# Aplicar K-Means con k=3
kmeans = KMeans(n_clusters=3, random_state=42)
df['cluster'] = kmeans.fit_predict(df_scaled)

# Asignar nombres a los clusters
cluster_names = {0: 'Líder del mercado', 1: 'Fidelidad garantizada', 2: 'Exploradores frecuentes'}
df['cluster_name'] = df['cluster'].map(cluster_names)

# Graficar los clusters (usando dos variables, por ejemplo, 'edad' y 'ingresos')
plt.figure(figsize=(8, 6))
sns.scatterplot(x=df['edad'], y=df['ingresos'], hue=df['cluster'], palette='viridis')
plt.title('Clusters de Clientes (Edad vs Ingresos)')
plt.xlabel('Edad')
plt.ylabel('Ingresos')

# Añadir nombres a cada cluster en el gráfico
for i, name in cluster_names.items():
    cluster_data = df[df['cluster'] == i]
    plt.text(cluster_data['edad'].mean(), cluster_data['ingresos'].mean(), name,
             fontsize=12, ha='center', color='black')

plt.show()
```

P8. Predicción de «High Margin»

```python
# Asegurarnos de que las fechas estén en formato datetime
df['fecha_ultima_compra'] = pd.to_datetime(df['fecha_ultima_compra'])

# Calcular la frecuencia mensual por cliente
def calcular_frecuencia_mensual(fecha_compra):
    # Si el cliente tiene solo una compra, asignamos la frecuencia como 1 compra por mes
    if fecha_compra.nunique() == 1:
        return 1
    else:
        # Número de compras únicas / meses
        return fecha_compra.nunique() / ((fecha_compra.max() - fecha_compra.min()).days / 30)

df['frecuencia_mensual'] = df.groupby('id_cliente')['fecha_ultima_compra'].transform(calcular_frecuencia_mensual)

# Calcular el promedio del ticket
df['promedio_ticket'] = df['total_gastado'] / df['compras']

# Calcular la tasa de retención según el intervalo de días desde la última compra
def calcular_tasa_retencion(dias):
    if dias <= 90:
        return 0.80
    elif 91 <= dias <= 180:
        return 0.60
    else:
        return 0.30

# Calcular los días desde la última compra para cada cliente
df['dias_desde_ultima_compra'] = (df['fecha_ultima_compra'].max() - df['fecha_ultima_compra']).dt.days

# Asignar la tasa de retención
df['tasa_de_retencion'] = df['dias_desde_ultima_compra'].apply(calcular_tasa_retencion)

# Calcular LTV
df['LTV'] = df['promedio_ticket'] * df['frecuencia_mensual'] * 12 * df['tasa_de_retencion']

# Calcular el margen
df['margen'] = df['LTV'] - df['costo_adquisicion']

# Calcular el percentil 75 de la columna 'margen'
percentil_75 = df['margen'].quantile(0.75)

# Crear la nueva columna 'high_margin' donde 1 es si margen >= percentil 75, de lo contrario 0
df['high_margin'] = (df['margen'] >= percentil_75).astype(int)

# Verificar los primeros valores
print(df[['margen', 'high_margin']].head())
```

**Decisión Estratégica**

Asignación de presupuesto (US$10 000)
Simula 3 escenarios de inversión:
i) Concentrar todo en un solo cluster.
ii) Distribuir en dos clusters clave.
iii) Asignar proporcional al margen esperado.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Crear un DataFrame con ejemplos de clusters
df_clusters = pd.DataFrame({
    'cluster': ['Cluster 1', 'Cluster 2', 'Cluster 3'],
    'margen': [5000, 12000, 8000],  # márgenes esperados por cluster
    'costo_adquisicion': [2000, 3000, 2500]  # costos de adquisición por cluster
})

# Calcular el ROI para cada cluster
df_clusters['ROI'] = (df_clusters['margen'] - df_clusters['costo_adquisicion']) / df_clusters['costo_adquisicion']

# Escenario i) Concentrar todo en el cluster con el mayor ROI
cluster_max_roi = df_clusters.loc[df_clusters['ROI'].idxmax()]
roi_concentrado = cluster_max_roi['ROI']

# Escenario ii) Distribuir en dos clusters clave (los dos con mayor ROI)
top_2_clusters = df_clusters.nlargest(2, 'ROI')
roi_distribuido = top_2_clusters['ROI'].mean()

# Escenario iii) Asignar proporcional al margen esperado
total_margen = df_clusters['margen'].sum()
df_clusters['porcentaje_margen'] = df_clusters['margen'] / total_margen
presupuesto_asignado = df_clusters['porcentaje_margen'] * 10000
df_clusters['ROI_ponderado'] = (presupuesto_asignado - df_clusters['costo_adquisicion']) / df_clusters['costo_adquisicion']
roi_ponderado = df_clusters['ROI_ponderado'].sum()

# Resultados
resultados = pd.DataFrame({
    'Escenario': ['Concentrado', 'Distribuido en 2', 'Proporcional al margen'],
    'ROI Esperado': [roi_concentrado, roi_distribuido, roi_ponderado]
})

# Mostrar resultados
print(resultados)

# Gráfico Tornado (Comparación de ROI Esperado)
resultados.set_index('Escenario')['ROI Esperado'].plot(kind='barh', color=['red', 'green', 'blue'])
plt.title('Comparación de ROI Esperado por Escenario')
plt.xlabel('ROI Esperado')
plt.ylabel('Escenarios')
plt.show()
```

Programa de referidos

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator

# Crear un DataFrame de ejemplo con clusters, recomendación y LTV
df_clients = pd.DataFrame({
    'id_cliente': range(1, 101),
    'cluster': np.random.choice(['Cluster 1', 'Cluster 2', 'Cluster 3'], size=100),
    'recomendaciones': np.random.uniform(0, 1, size=100),  # Recomendación aleatoria entre 0 y 1
    'LTV': np.random.uniform(1000, 5000, size=100)  # LTV aleatorio
})

# Calcular el top 10% de advocates basados en las recomendaciones
top_10_advocates = df_clients.nlargest(int(0.1 * len(df_clients)), 'recomendaciones')

# Promedio del LTV por cluster
avg_LTV_by_cluster = df_clients.groupby('cluster')['LTV'].mean()

# Estimar el valor futuro
top_10_advocates['valor_futuro'] = 2 * avg_LTV_by_cluster[top_10_advocates['cluster']].values

# Crear una columna con el cambio en el valor futuro para el gráfico de cascada
top_10_advocates['cambio_valor_futuro'] = top_10_advocates['valor_futuro'] - top_10_advocates['LTV']

# Resultados generales
total_inicial = top_10_advocates['LTV'].sum()
total_futuro = top_10_advocates['valor_futuro'].sum()

# Preparar los datos para el gráfico de cascada
cambio_acumulado = np.cumsum(top_10_advocates['cambio_valor_futuro'])

# Graficar el gráfico de cascada
fig, ax = plt.subplots(figsize=(10, 6))
ax.bar(np.arange(len(top_10_advocates)), cambio_acumulado, color='royalblue', edgecolor='black')
ax.axhline(0, color='black', linewidth=1)
ax.set_title("Gráfico de Cascada del Valor Futuro de los Top 10% de Advocates")
ax.set_xlabel("Advocates")
ax.set_ylabel("Cambio en el Valor Futuro")
ax.xaxis.set_major_locator(MaxNLocator(integer=True))

plt.show()

# Mostrar resultados
print(f"Valor inicial total: {total_inicial:.2f}")
print(f"Valor futuro total: {total_futuro:.2f}")
```

P11. Sesgo y limitaciones


**Posibles Sesgos del Dataset:**

1. Sesgo por Comportamiento:
Si solo se consideran los clientes que ya han interactuado mucho con la plataforma, podríamos estar perdiendo información de los clientes que no han tenido muchas interacciones pero aún tienen potencial.

2. Sesgo por Tiempo:
Si el dataset se concentra en un periodo corto, no refleja el comportamiento de los clientes a lo largo del tiempo, lo que podría llevar a conclusiones erróneas sobre su valor.

3. Sesgo de Reporte:
Si hay errores al registrar la información, como omitir algunas compras o interacciones, el análisis de los datos podría estar sesgado y no ser completamente preciso.

**Mejoras Metodológicas:**

1. Segmentación Más Precisa:
En lugar de usar categorías amplias, se podrían crear segmentos más específicos de clientes (por ejemplo, por tipo de compra o frecuencia) para obtener análisis más detallados y recomendaciones más personalizadas.

2. Incorporar Fuentes de Datos Externas:
Usar información adicional, como datos del comportamiento en redes sociales o interacciones en otros canales, podría ayudar a mejorar la comprensión del cliente y predecir su valor con mayor precisión.
